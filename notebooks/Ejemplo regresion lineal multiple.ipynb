```python
def tidy_corr_matrix(corr_mat):
    '''
    Función para convertir una matriz de correlación de pandas en formato tidy
    '''
    corr_mat = corr_mat.stack().reset_index()
    corr_mat.columns = ['variable_1','variable_2','r']
    corr_mat = corr_mat.loc[corr_mat['variable_1'] != corr_mat['variable_2'], :]
    corr_mat['abs_r'] = np.abs(corr_mat['r'])
    corr_mat = corr_mat.sort_values('abs_r', ascending=False)

    return(corr_mat)


corr_matrix = datos.select_dtypes(include=['float64', 'int']).corr(method='pearson')
tidy_corr_matrix(corr_matrix).head(10)
```

```python
#Matriz de correlacion
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(4, 4))

sns.heatmap(
    corr_matrix,
    annot     = True,
    cbar      = False,
    annot_kws = {"size": 8},
    vmin      = -1,
    vmax      = 1,
    center    = 0,
    cmap      = sns.diverging_palette(20, 220, n=200),
    square    = True,
    ax        = ax
)

ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation = 45,
    horizontalalignment = 'right',
)

ax.tick_params(labelsize = 10)
```

#Heterocedasticidad:
De esta forma, la heterocedasticidad es una propiedad estadística que ocurre cuando la variabilidad de un error aleatorio, o residuo, es inconsistente a lo largo de todas las observaciones en un conjunto de datos. O dicho de otro modo, el error vislumbrado no es persistente, si no caótico, hay dispersión o variabilidad de la variable dependiente que cambia a lo largo del conjunto de datos

## Test de heterocedasticidad:
### Normalidad en los residuos (test de shapiro)

Cuando hacemos este test, estamos realizando una prueba de hipotesis de que los residuos tienen distribucion normal, analizando el p-value

H0 = Dist. normal

El p-value no tiene que ser mayor que el estadistico

```python
shapiro_test = stats.shapiro(residuos_train)
shapiro_test
#return (stadistico,p-value)
# como el p-value resultante no es mayor que el estadistico, aceptamos la hipotesis de que los errores se distribuyen normal

```

### Normalidad de los residuos D'Agostino's K-squared test



```python
k2, p_value = stats.normaltest(residuos_train)
print(f"Estadítico= {k2}, p-value = {p_value}")
```
